```python
pyautogui.click(1444,40)
time.sleep(2)
pyautogui.moveTo(446,369,0.3)

distance = 400
change = 40

while distance>0:
    time.sleep(0.2)
    pyautogui.drag(distance,0)
    pyautogui.drag(0,distance)
    distance -= change
    pyautogui.drag(-distance,0)
    pyautogui.drag(0,-distance)
    distance -= change
```
